```
# Data
data = {
    'salary' : [1000,2000,3000,4000,5000,2000,8000,9000,7000,10000,1000,2000,3000,4000,5000,2000,8000,9000,7000,10000],
}
```

```
df = pd.DataFrame(data)
df.head()
```

```
df.shape
```

```
# Line - Plot
plt.plot(df['salary'],color='red',marker='o',linestyle=':',linewidth="2")
plt.grid()
plt.show()
```

```
# Histo-gram chart
plt.hist(df['salary'])
plt.grid()
plt.show()
```

```
plt.hist(df['salary'],bins=5,color="green")
plt.grid()
plt.show()
```

```
#boxplot
plt.boxplot(df['salary'])
plt.grid()
plt.show()
```

```
df.loc[20]=[-8000]
```

```
plt.boxplot(df['salary'])
plt.grid()
plt.show()
```

```
#univariate : Categorical
df['dept']= ['HR','IT','HR','Sales','Finance','HR','IT','HR','Sales','Finance','HR','IT','HR','Sales','Finance','HR','IT','HR','Sales','HR','Sales']
```

```
df.head()
```

```
df.info()
```

```
#pie chart
count = df['dept'].value_counts()
count
```

```
plt.pie(count,labels=count.index,autopct="%1.2f",explode=[0,0.1,0.2,0.3])
plt.axis("equal") #for get complete circle if you look like pie chart is noy in perfect circle
plt.show()
```

```
#count plot
plt.bar(count.index,count,color=["green","red","grey","purple"])
plt.show()
```

```
#bivariate - numerical - numerical
df.head()
```

```
df['age'] = [25,24,26,24,20,30,21,22,25,24,26,24,20,30,21,22,25,24,26,24,20]
```

```
df.head()
```

```
plt.scatter(df['salary'],df['age'])
plt.grid()
plt.show()
```

```
sort_age = sorted(df['age'])
```

```
#line chart-2
plt.plot(sort_age,df['salary'],color='red',marker="*",linestyle=":",linewidth="3")
plt.grid()
plt.show()
```

```
#bar chart
plt.bar(df['age'],df['salary'],color='green',align="edge")
plt.grid()
plt.show()
```

```
# bivariant : numerical - categorical
HR_salary = df[df['dept']=='HR']['salary']
HR_salary
```

```
IT_sal = df[df['dept']=='IT']['salary']
IT_sal
```

```
fin_sal = df[df['dept']=='Finance']['salary']
fin_sal
```

```
sales_sal = df[df['dept']=='Sales']['salary']
sales_sal
```

```
#box plot
plt.boxplot([HR_salary,IT_sal,fin_sal,sales_sal],labels =["Hr","IT","fin","sale"])
plt.legend()
plt.grid()
plt.show()
```

```
sal_by_dpt=df.groupby('dept')['salary'].sum()
sal_by_dpt
```

```
#pie chart
plt.pie(sal_by_dpt,labels= sal_by_dpt.index,autopct="%1.2f")
plt.show()
```

```
hr_mean = sum(HR_salary)/len(HR_salary)
fin_mean = sum(fin_sal)/len(fin_sal)
it_mean = sum(IT_sal)/len(IT_sal)
sales_mean = sum(sales_sal)/len(sales_sal)
```

```
# bar chart
plt.bar(["Hr", "fin","it","sal"],[hr_mean,fin_mean,it_mean,sales_mean],color=["green","black","red","yellow"])
plt.show()
```

```
#multi-varient analysis
df['exp']= [1,2,2.5,10,3.5,9,8,7.5,6,5.5,2.5,10,3.5,9,8,7.5,6,5.5,2.5,4,6]
df.head()
```

```
# bubble plot
plt.scatter(df['salary'],df['age'],s=df['exp']*20,color="skyblue",edgecolor="black")
plt.title("Age VS salary VS Experience")
plt.xlabel("Age")
plt.ylabel("Salary")
plt.grid()
plt.show()
```

```
#two numerical and 1 categorical
plt.scatter(df['age'],df['salary'],c=df['dept'].map({"HR" : "red", "IT":"yellow","Finance":"orange","Sales":"blue"}))
plt.title("Age VS Dept VS Salary")
plt.xlabel("age")
plt.ylabel("Salary")
plt.legend()
plt.show()
```

```
#object oriented Api
```

```
fig , axs = plt.subplots(1,3,figsize=(10,5))
#line plot
axs[0].plot(sort_age,df['salary'],color='red',marker="*",linestyle=":",linewidth="3")
axs[0].grid()
axs[0].set_title("Salery VS age")
axs[0].set_xlabel("age")
axs[0].set_ylabel("salary")

#histogram
axs[1].hist(df['salary'],bins= 5 , color="green")
axs[1].grid()
axs[1].set_title("Histogram")
axs[1].set_xlabel("salary")
axs[1].set_ylabel("frequency")

#boxplot
axs[2].boxplot(df['salary'])
axs[2].grid()
axs[2].set_title("Boxplot")
axs[2].set_xlabel("salary")
axs[2].set_ylabel("frequency")

plt.savefig("multipleplots.png")

plt.show()
```
