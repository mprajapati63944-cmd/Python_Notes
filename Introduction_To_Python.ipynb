#COMMENTS

comments are used to explain code and are completely ignored by the interpreter during program execution. They help enhance readability, provide context for complex logic, and can be used for debugging by temporarily disabling code

##Types
Single-line : Single-line comments start with the hash symbol (#). All text following the # to the end of the line is considered a comment

Inline : this comment is created using the hash symbol (#) placed after a statement on the same line. The Python interpreter ignores the # symbol and everything that follows it until the end of the line

Multi-line : Python does not really have a syntax for multiline comments.
To add a multiline comment you could insert a # for each line
Or, not quite as intended, you can use a multiline string.

Since Python will ignore string literals that are not assigned to a variable, you can add a multiline string (triple quotes) in your code, and place your comment inside it


```
# This is a single line comment
print("hello world") #this is inline comment
'''
this is
multi-line
comment
'''
print("thats all about comment")
```

#Type Conversion
Python type conversion is the process of changing a value from one data type to another
 Implicit : automatically by the interpreter
 Explicit : manually by the programmer using built-in functions

```
#implicit
num_int = 10     # integer
num_float = 5.5  # float

result = num_int + num_float

print(result)
print(type(result))
```

```
#explicit
num_string = "100" # string
num_int = 50       # integer

# Convert string to integer using the int() function
converted_num = int(num_string)

result = num_int + converted_num

print(result)
print(type(result))
```

```
#Arthematic Operators
# Variables
a = 15
b = 4

# Addition
print("Addition:", a + b)

# Subtraction
print("Subtraction:", a - b)

# Multiplication
print("Multiplication:", a * b)

# Division
print("Division:", a / b)

# Floor Division
print("Floor Division:", a // b)

# Modulus
print("Modulus:", a % b)

# Exponentiation
print("Exponentiation:", a ** b)
```
