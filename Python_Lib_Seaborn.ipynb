#Seaborn


```
import pandas as pd
import numpy as np
import seaborn as sns
```

```
sns.get_dataset_names()
```

```
df= sns.load_dataset('penguins')
df.head()
```

```
df['species'].value_counts()
```

```
df['island'].value_counts()
```

```
sns.set_style('ticks')
sns.set_context('talk')
sns.scatterplot(data=df,x='flipper_length_mm',y='body_mass_g',hue='island',palette='Set2')
sns.despine()
```

```
#scatterplot
sns.set_context('notebook')
sns.scatterplot(data=df,x='species',y='body_mass_g',hue='island',style='sex',palette='rocket',alpha=0.5)
```

```
#stripplot
sns.stripplot(data=df,x='species',y='body_mass_g',hue='island')
```

```
#swarnplot
sns.swarmplot(data=df,x='species',y='body_mass_g',hue='island')
sns.set_context('talk')
```

```
#histogram
sns.histplot(data=df,x='body_mass_g',hue='sex',multiple="stack")
```

```
#regression plot
sns.regplot(data=df,x='body_mass_g',y='flipper_length_mm',color='green')
```

```
#line plot
sns.set_style('whitegrid')
sns.set_context('notebook')
sns.lineplot(data=df,y='flipper_length_mm',x='body_mass_g',hue='island',style='sex')
```

```
#joint plot
sns.jointplot(data=df,x='body_mass_g',y='flipper_length_mm',hue='sex',kind='kde')
```

```
#joint plot
sns.jointplot(data=df,x='body_mass_g',y='flipper_length_mm',kind='reg')
```

```
#bar plot
sns.barplot(data=df,x='species',y='body_mass_g',hue='sex',palette=['skyblue','pink'],estimator=np.median)
```

```
#count plot
sns.countplot(data=df,x='species',hue='island',palette='Set2')
```

```
#box plot
sns.boxplot(data=df,x='species',y='body_mass_g',hue='sex',palette='Set2')
```

```
#violen plot
sns.violinplot(data=df,x='species',y='body_mass_g',hue='sex',palette='Set2',split=True)
```

```
#swarn plot with violen plot
sns.violinplot(data=df,x='species',y='body_mass_g',palette='pastel')
sns.swarmplot(data=df,x='species',y='body_mass_g',hue='island',size=3,palette='dark')
#
```

```
#kde plot
sns.kdeplot(data=df,x='body_mass_g',hue='species',palette='pastel',fill=True)
```

```
df.head()
```

```
#heat map
columns = ['bill_length_mm','bill_depth_mm','flipper_length_mm','body_mass_g']
df[columns].corr()
```

```
sns.heatmap(data=df[columns].corr(),annot=True,cmap='coolwarm',vmin=-0.2,linewidths=2,linecolor='black')
```

```
#rug plot
sns.rugplot(data=df,y='body_mass_g',hue='species',height=0.2)
```

```
#pair plot
sns.pairplot(data=df,hue='sex',palette='Set2',diag_kind='kde')
```

```
#pair plot
sns.pairplot(data=df,hue='sex',palette='Set2',diag_kind='hist')
```

```
#pair grid
graph = sns.PairGrid(data=df,hue='sex',palette='Set2')
graph.map_diag(sns.histplot)
graph.map_upper(sns.scatterplot)
graph.map_lower(sns.kdeplot)
graph.add_legend()
```
