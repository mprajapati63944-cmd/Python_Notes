#PANDAS LIBRARY

Pandas is an essential open-source Python library designed for data analysis and manipulation. It provides high-performance, easy-to-use data structures, most notably the Series and DataFrame, which simplify working with tabular data.


```
import pandas as pd
print(pd.__version__)
```

```
#series
data = pd.Series(["BMW","volvo","BMW"])
data
```

```
colors = pd.Series(["red","blue","yellow","green"])
colors
```

```
mydataset = {
  'cars': ["BMW", "Volvo", "Ford","BMW", "Volvo", "Ford"],
  'passings': [3, 7, 2,3, 7, 2]
}
myvar = pd.DataFrame(mydataset)
myvar
```

```
#make a series of different foods
foods = pd.Series(["burgur","pizza","coldrinks","softdrinks","manchurian","chilli potato"])
foods
```

```
# make a series of different dollar values (in integer)
prices = pd.Series([99,299,85,80,240,120])
prices
```

```
# combine the series of foods and price in a dataframe
data = pd.DataFrame({"foods":foods,"prices":prices})
data
```

```
#importing Data
car_sales = {
    "make" : ["Toyota","honda","Toyota","BMW","Nissan","Toyota","Honda","Honda","Toyota","Nissan"],
    "color" : ["White","Red","Blue","Black","white","Green","Blue","Blue","White","White"],
    "Odometer" : [150043,87899,32549,11179,213095,99213,45698,54738,60000,31600],
    "Doors" : [4,4,3,5,4,4,4,4,4,4],
    "Price" : [4000.00,5000.00,7000.00,22000.00,3500.00,4500.00,7500.00,6000.00,6250.00,9700.00]

}
car_sales = pd.DataFrame(car_sales)
car_sales
file_name = 'Car_sales.csv'
car_sales.to_csv(file_name, index=False)
```
