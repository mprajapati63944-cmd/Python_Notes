#Data Cleaning
## 1. Handeling Missing Vaues :
-> Identify missing values (often represented as NaN or None) using df.isnull().sum() to count them per column.

-> Remove rows or columns with missing data using df.dropna().

-> Fill missing values using df.fillna() with a specified value (e.g., zero, mean, median, or a new category).

-> You can fill a specific column using df['column_name'].fillna(value).

```python
import pandas as pd #import Pyhon
data = pd.read_csv('/content/Titanic-Dataset.csv') #load dataset
```

```python
data.head()  #get first 5 rows
```

```python
#check null values
data.isna().sum()
```

```python
#handling null values in age by median of the age
data['Age'].fillna(data['Age'].median(), inplace=True)
data.isna().sum() #again check null values
```

```python
#fill null values of cabin
data['Cabin'].fillna('Unknown', inplace=True)
data.isna().sum() #again check null values
```

```python
#fill the null values of Embarked
data['Embarked'].fillna(data['Embarked'].mode()[0], inplace=True)
data.isna().sum() #again check null values
```

```python
#data after handeled null values
data.head()
```

##Manage Outliers and Wrong Data:
-> Identify outliers using statistical methods (e.g., IQR method with quantile()) or visualizations (e.g., box plots).

-> Handle them by removing the rows, replacing them with a more reasonable value, or using methods that are robust to outliers.

```python
#checking the outliers
Q1 = data['Age'].quantile(0.25)
Q3 = data['Age'].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
outliers = data[(data['Age'] < lower_bound) | (data['Age'] > upper_bound)]
outliers #get outliers
```

```python
#replace outliers by median
data.loc[data['Age'] < lower_bound, 'Age'] = data['Age'].median()
data.loc[data['Age'] > upper_bound, 'Age'] = data['Age'].median()
data
```
