#List

```python
#Task 19:  Implementation of  removing an element by index
names=['parth', 'mohit', 'chetan', 'prajapati', 'pathak', 'singh']
if len(names) > 5:
  del(names[5])

print(names)
```

#SETS

```python
#Task 1: Implementation of Creating Sets
set1={1,2,3,4,5}
print(set1)
```

```python
#Task 2: Implementation of Adding and Removing Elements in a Set
set1={1,2,3,4,5}
set1.add(6)
set1.remove(5)
print(set1)
```

```python
#Task 3: Implementation of Union of Sets
set2={6,7,8,9,10}
union_set=set1.union(set2)
print(union_set)
```

```python
#Task 4: Implementation of Intersection of Sets
intersection_set=set1.intersection(set2)
print(intersection_set)
```

```python
#Task 5: Implementation of Difference of Sets
set1={1,2,3,4,5,6,7}
set2={6,7,8,9,10}
difference_set=set1.difference(set2)
print(difference_set)
```

```python
#Task 6: Implementation of Symmetric Difference of Sets
symmetric_difference_set=set1.symmetric_difference(set2)
print(symmetric_difference_set)
```

```python
#Task 7: Implementation of Subset and Superset Testing
set3={1,2,3,4}
set4={1,2,3,4,5,6}
print(set3.issubset(set4))
```

```python
#Task 8: Implementation of Set Membership Testing
print(3 in set1)
```

```python
#Task 9: Implementation of Removing Duplicates Using Sets
numbers=[1,1,3,5,1,2,4,8,6,7]
numbers=set(numbers)
print(numbers)
```

```python
#Task 10: Implementation of Set Comprehension
square_set={x**2 for x in range(1,6)}
print(square_set)
```

```python
#Task 11: Implementation of Clearing a Set
set1={1,5,2,4,7,8,9}
print(set1)
set1.clear()
print(set1)
```

#DICTIONARY

```python
#Task 1: Implementation of Creating a Dictionary
student_grades = {
    "mohit" : 8.4,
    "ayush" : 8.03,
    "parth" : 8.1
}
print(student_grades)
```

```python
#Task 2: Implementation of Accessing Dictionary Values
print(student_grades["mohit"])
```

```python
#Task 3: Implementation of Adding a New Key-Value Pair
student_grades["shreya"]=8.5
print(student_grades)
```

```python
#Task 4: Implementation of Removing a Key-Value Pair
student_grades = {
    "mohit" : 8.4,
    "ayush" : 8.03,
    "parth" : 8.1
}
student_grades.pop("ayush")
print(student_grades)
```

```python
#Task 5: Implementation of Checking Key Membership
print("mohit" in student_grades)
```

```python
#Task 6: Implementation of Iterating Over a Dictionary (Keys and Values)
student_grades = {
    "mohit" : 8.4,
    "ayush" : 8.03,
    "parth" : 8.1
}
for student, grade in student_grades.items():
  print(f"{student}:{grade}")
```

```python
#Task 7: Implementation of Using the get() Method
print(student_grades.get("mohit","not found"))
```

```python
#Task 8: Implementation of Merging Dictionaries
student_grades2 = {
    "shreya" : 8.5,
    "chetan" : 8.6
}
student_grades.update(student_grades2)
print(student_grades)
```

```python
#Task 9: Implementation of Dictionary Comprehension
student_grades = {x:x**2 for x in range(1,6)}
print(student_grades)
```

```python
#Task 10: Implementation of Handling Nested Dictionaries
student_grades= {
    "mohit" : {
        "grade" : 8.4,
        "class" : "A"
    },
    "ayush" : {
        "grade" : 8.03,
        "class" : "B"
    }
}
print(student_grades)

print(student_grades['mohit']["grade"])
```

```python
#Task 11: Implementation of Clearing and Copying Dictionaries
student_grades= {
    "mohit" : {
        "grade" : 8.4,
        "class" : "A"
    },
    "ayush" : {
        "grade" : 8.03,
        "class" : "B"
    }
}
student_grades_copy=student_grades.copy()
student_grades.clear()
print(student_grades)
print(student_grades_copy)
```

#Strings

```python
#Task 1: Implementation of Creating and Accessing Strings
word="Mohit Prajapati"
print(word)
```

```python
#Task 2: Implementation of String Concatenation and Repetition
word1="Mohit"
word2="Prajapati"
print(word1+" "+word2)
```

```python
#Task 3: Implementation of String Case Manipulation
print(word.upper())
print(word.lower())
print(word.title())
print(word.swapcase())
```

```python
#Task 4: Implementation of Searching in Strings
word="mohit prajapati"
print(word.find("prajapati"))
```

```python
#Task 5: Implementation of Replacing Substrings
word="mohit prajapati"
print(word.replace("mohit","himanshu"))
```

```python
#Task 6: Implementation of String Formatting
name="mohit"
age=19
string=f"I'm {name} and I'm {age} years old"
print(string)
```

```python
#Task 7: Implementation of  Trimming and Padding String
word="    mohit prajapati     "
print(word.strip())
```

```python
#Task 8: Implementation of  Splitting and Joining Strings
sentence="mohit prajapati"
print(sentence.split())
print('-'.join(sentence))
```

```python
#Task 9: Implementation of  Counting Characters
word="hi this is mohit prajapati"
print(word.count('i'))
```

```python
#Task 10: Implementation of  Checking String Properties
is_alnum ="mohit12".isalnum()
is_alpha = "mohit".isalpha()
is_digit = "123456".isdigit()
is_space = "   ".isspace()
print(is_alpha,is_alnum,is_digit,is_space,sep="\n")
```
