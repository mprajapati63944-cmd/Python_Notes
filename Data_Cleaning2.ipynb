#Data cleaning on two datas


```
import pandas as pd
#import pandas
```

```
#import first data
data1=pd.read_excel("/content/2_HR_Cleaning_Practice.xlsx")
```

```
#get first 5 rows
data1.head()
```

```
#get last 5 rows
data1.tail()
```

##Handelling Null values
-> find null values

-> fill or delete null values rows

```
#find null values
data1.isnull().sum()
```

```
#we see the Years_exp column have 326 null values
#we can replace the null values by 0
#means the the employee have 0 years of experience
data1['Years_Exp'].fillna(0,inplace=True)
```

```
#check values are filled or not
data1.isnull().sum()
```

```
#null values are 0 means all null value are handelled
```

#Handelling Duplicates values
-> find duplicates

-> delete them

```
#check duplicate values
data1.duplicated().sum()
```

```
#there are 600 duplicate rows in data
#now delete the duplicates
data1.drop_duplicates(inplace=True)
```

```
#checking duplicates are deleted or not
data1.duplicated().sum()
```

```
#checking rows and columns
data1.shape
```

```
#check data again and do staistic calculations
data1.head()
```

```
data1.describe()
```

#Data cleaning for second data
